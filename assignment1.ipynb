```python
import math
import cmath

import matplotlib.pyplot as plt
import pandas as pd
```

```python
complex_numbers = [
    complex(1,1), 
    complex(2,1), 
    complex(6,3), 
    complex(-3,4), 
    complex(6,-1)
]
sum(complex_numbers)
```

```python
plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(15, 15))
plt.scatter([c.real for c in complex_numbers], [c.imag for c in complex_numbers])
plt.title("$\mathbb{C}$ plane", size=30)
plt.xlabel("Real")
plt.ylabel("Imaginary")
```
